**Primero:** extraer audio del video y guardar en archivo .wav

```python
import ffmpeg

# Ruta del archivo de video.
archivo_video = "/content/Meadow - How To Play - board game rules.mp4"
audio_video = ffmpeg.input(archivo_video)

# Especifica la salida como audio en formato WAV.
audio = ffmpeg.output(audio_video, 'audio_english.wav', f='wav')

# Instead of passing the file object, read the contents into a bytes object
with open(archivo_video, 'rb') as input_file:
    video_data = input_file.read()  # Read the entire file into video_data

ffmpeg.run(audio, input=video_data, overwrite_output=True) # Pass video_data to ffmpeg.run
```

**Segundo**: el archivo es muy extenso, por lo que, previamente a extraer el texto, lo fragmentamos en segmentos de menor duracion.  

```python
from pydub import AudioSegment
from pydub.silence import split_on_silence

# Carga el archivo de audio
audio = AudioSegment.from_wav("audio_english.wav")

# Divide el audio en fragmentos basado en los silencios
fragments = split_on_silence(audio, min_silence_len=700, silence_thresh=-36)

# Guarda cada fragmento de audio
for i, fragment in enumerate(fragments):
    fragment.export(f"segment_{i}.wav", format="wav")
    print(f"Fragmento {i} guardado como segment_{i}.wav")
```

**Tercero:** se carga y extrae el texto de cada segmento y se guarda el texto completo en un archivo .txt

```python
import speech_recognition as sr

# Crea un objeto Recognizer
r = sr.Recognizer()
all_text = ""

# Abre el archivo de audio
for i in range(37):
  try:
    with sr.AudioFile(f"segment_{i}.wav") as source:
        audio_data = r.record(source)
        texto = r.recognize_google(audio_data, language="en-US")
        all_text += texto + " "
  except Exception as e:
    print(f"Error en el archivo segment_{i}.wav: {e}")

# Imprime el texto transcrito
print(all_text)
# Guarda el texto en un archivo
with open('Pradera_Tutorial_English_video.txt', 'w', encoding='utf-8') as archivo:
    archivo.write(all_text)
```

### Extraer texto de documentos PDF

```python
!pip install PyPDF2
```

```python
from PyPDF2 import PdfReader
```

```python
# Abre el archivo en modo binario de lectura ('rb')
with open('/content/Pradera_Reglamento_(Spanish).pdf', 'rb') as archivo:

  # Crea un objeto PdfFileReader
  lector = PdfReader(archivo)

  # Inicializa una cadena vacía para almacenar el texto
  texto = ''

  # Itera sobre todas las páginas del PDF
  for i in range(len(lector.pages)):
     # Obtiene la página
      pagina = lector.pages[i]
     # Extrae el texto de la página y lo añade a la cadena de texto
      texto += pagina.extract_text ()
# Imprime el texto extraído
  print(texto)
  with open('reglamento_español.txt', 'w', encoding='utf-8') as archivo:
    archivo.write(texto)
```

```python
# PODEMOS VER LA METADATA DEL ARCHIVO PDF
reader = PdfReader('/content/Pradera_Reglamento_(Spanish).pdf')
meta = reader.metadata

print("Total Pages: ", len(reader.pages))

print("Author: ", meta.author)
print("Creator: ", meta.creator)
print("Producer: ", meta.producer)
print("Subject: ", meta.subject)
print("Title: ", meta.title)
```
